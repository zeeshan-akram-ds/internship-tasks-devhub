# 3. Visual Exploration

### Scatter Plot to Analyze Relationships Between Variables

```python
## Lets plot scatter plot to analyze relationship between sepal_length and sepal_width with color-code the target column for better understanding
```

```python
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', style='species', s=70)
plt.title("Sepal Length vs Sepal Width by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species")
plt.tight_layout()
plt.show()
```

### The scatter plot reveals a separation between setosa and the other two species in terms of sepal measurements. However, versicolor and virginica show some overlap, indicating these features alone might not fully differentiate them.

### Histogramt to Examine Data Distribution

```python
# Histogram showing the distribution of petal length by species
sns.histplot(data=df, x='petal_length', hue='species', kde=True, palette='muted', edgecolor='black', bins=20)
plt.title("Distribution of Petal Length Across Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

### The histogram reveals that setosa flowers have consistently shorter petal lengths, while virginica shows a wider spread with longer petals. The overlap between versicolor and virginica suggests that petal length could be a strong feature for differentiating species, especially setosa, but may require support from other features for full class separation.

# 3. Box Plot to Detect Outliers and Value Spread

```python
# Box plot to visualize the distribution and outliers of petal width across species
sns.boxplot(hue='species', y='petal_width', data=df, palette='pastel')
plt.title("Distribution of Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.tight_layout()
plt.show()
```

### This box plot highlights clear differences in petal width among the three species. setosa has consistently low petal width with no overlap but shows some outliers, while virginica shows the highest range. These differences are valuable for classification tasks, indicating that petal width is a strong distinguishing feature.
